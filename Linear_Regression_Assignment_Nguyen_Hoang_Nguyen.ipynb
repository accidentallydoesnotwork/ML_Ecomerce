## Training and Testing Data

**8.** Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.
** Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column. **
5pts

```python
# x = customers[['Avg. Session Length'],['Time on App'],['Time on Website'],['Length of Membership']]
customers_1 = customers.drop(['Email', 'Address', 'Avatar','Yearly Amount Spent'], axis=1)
x = customers_1
y = customers['Yearly Amount Spent']
```

**9. Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set test_size=0.3 and random_state=101**
5pts

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
x_train = x_train.sort_index()
y_train = y_train.sort_index()
```

## Training the Model

Now its time to train our model on our training data!

**10. Import LinearRegression from sklearn.linear_model **
5pts

```python
from sklearn.linear_model import LinearRegression
```

**11. Create an instance of a LinearRegression() model named lm.**
5pts

```python
lm = LinearRegression()
```

**12. Train/fit lm on the training data.**
5pts

```python
lm.fit(x_train, y_train)
```

**13. Print out the coefficients of the model**
5pts

```python
# Hint: coef_
lm.coef_
```

## Predicting Test Data
Now that we have fit our model, let's evaluate its performance by predicting off the test values!

**14. Use lm.predict() to predict off the X_test set of the data.**
5pts

```python
y_predicted = lm.predict(x_test)
```

**15. Create a scatterplot of the real test values versus the predicted values. **
5pts

```python
plt.scatter(x = y_test, y = y_predicted,)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y') 
```

## Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares.

**16. Calculate: Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error**
10pts

Note: Refer to those links for the formulas

[Mean Absolute Error(MAE)](https://en.wikipedia.org/wiki/Mean_absolute_error)

[Mean Squared Error(MSE)](https://en.wikipedia.org/wiki/Mean_squared_error)

[Root Mean Squared Error(RMSE)]()

Reference

[MAE and RMSE](https://medium.com/human-in-a-machine-world/mae-and-rmse-which-metric-is-better-e60ac3bde13d)

```python
import numpy as np
mse = np.mean(np.square(y_predicted - y_test))
mae = np.mean(np.absolute((y_test - y_predicted)))
rmse = np.sqrt(np.mean(np.square(y_predicted - y_test)))

print('MAE:', mae)
print('MSE:', mse)
print('RMSE:', rmse)
```

**19. How can you interpret these coefficients? **
10pts

Answer here: We can interpret the coefficients as how much the dependant variable ('Yearly Amount Spent') is affected when we increase a unit of each independent variable. For example: if the coefficient of Avg. Session Length is 25.98, we can say that if a Avg. Session Length variable increase by X unit, then the Yearly Amount spent variable will increase by 25.98 * X unit and vice versa. The same conclusion can also be reached for others type of variables: if Time on App variable increase by 1 unit, then the Yearly Amount Spent will increase by 38.59 units, so on and so forth. Therefore we can say that the Length of Membership variable affects the Yearly Amount Spent variable the most, since it has the highest coefficient.
